```python
### Default Parameter
params = {'Mass': 0.5,
             'kpPosXY': 1,
            'kpPosZ': 1,
          'KiPosZ': 20,
          'kpVelXY': 4,
          'kpVelZ': 4,
          'kpBank': 5,
          'kpYaw': 1,
          'kpPQR': [23, 23, 5]
         }
```

```python
### Make Parameters List

kpPosXY = [i for i in range(30, 33, 1)]  # 30 ~ 32
kpPosZ = [i for i in range(20, 51, 5)] # 20 ~ 50
KiPosZ = [i for i in range(30, 41, 5)] # 30 ~ 40
kpVelXY = [i for i in range(12, 14, 1)] # 12 ~ 13
kpVelZ = [i for i in range(9, 10, 1)]  # 9
kpBank = [i for i in range(7, 11, 1)] # 7 ~ 10
kpYaw = [i for i in range(2, 3, 1)] # 2
kpPQR = [[i, i, 6] for i in range(90, 96, 5)] # [90, 90, 6] ~ [95, 95, 6]
```

```python
listcase = []
for a in kpPosXY:
    for b in kpPosZ:
        for c in KiPosZ:
            for d in kpVelXY:
                for e in kpVelZ:
                    for f in kpBank:
                        for g in kpYaw:
                            for h in kpPQR:
                                listcase.append([a, b, c, d, e, f, g, h])
                                
print (len(listcase))
```

```python
kpPosXY = [i for i in range(30, 33, 1)]  # 30 ~ 32
kpPosZ = [i for i in range(20, 51, 5)] # 20 ~ 50
KiPosZ = [i for i in range(30, 41, 5)] # 30 ~ 40
kpVelXY = [i for i in range(12, 14, 1)] # 12 ~ 13

listcase2 = []
for a in kpPosXY:
    for b in kpPosZ:
        for c in KiPosZ:
            for d in kpVelXY:
                listcase2.append([a, b, c, d])

print (len(listcase2))
```

```python
### Save to Pickle
import pickle
with open('paramset.p', 'wb') as handle:
    pickle.dump(listcase, handle, protocol=pickle.HIGHEST_PROTOCOL)
```

```python
import ipywidgets

button = ipywidgets.Button(description="Button")
out = ipywidgets.Output()
cnt = 0

def on_button_clicked(b):
    global cnt
    #params['Mass'] = listcase[cnt][0]
    params['kpPosXY'] = listcase[cnt][0]
    params['kpPosZ'] = listcase[cnt][1]
    params['KiPosZ'] = listcase[cnt][2]
    params['kpVelXY'] = listcase[cnt][3]
    params['kpVelZ'] = listcase[cnt][4]
    params['kpBank'] = listcase[cnt][5]
    params['kpYaw'] = listcase[cnt][6]
    params['kpPQR'] = listcase[cnt][7]    
    print(listcase[cnt])
    SetQuadControlParams()
    
    cnt+=1
    
button.on_click(on_button_clicked)

display(button)
```

```python
# save the new paramter to QuadControlParams.txt
def SetQuadControlParams():
    f = open('../config/QuadControlParams.txt', 'r+')
    lines = []
    for line in f:
        if line.startswith('Mass ='):
            data = 'Mass = ' + str(params['Mass']) + '\n'
            lines = lines + [data]
        elif line.startswith('kpPosXY ='):
            data = 'kpPosXY = ' + str(params['kpPosXY']) + '\n'
            lines = lines + [data]
        elif line.startswith('kpPosZ ='):
            data = 'kpPosZ = ' + str(params['kpPosZ']) + '\n'
            lines = lines + [data]
        elif line.startswith('KiPosZ ='):
            data = 'KiPosZ = ' + str(params['KiPosZ']) + '\n'
            lines = lines + [data]    
        elif line.startswith('kpVelXY ='):
            data = 'kpVelXY = ' + str(params['kpVelXY']) + '\n'
            lines = lines + [data]    
        elif line.startswith('kpVelZ ='):
            data = 'kpVelZ = ' + str(params['kpVelZ']) + '\n'
            lines = lines + [data]    
        elif line.startswith('kpBank ='):
            data = 'kpBank = ' + str(params['kpBank']) + '\n'
            lines = lines + [data]    
        elif line.startswith('kpYaw ='):
            data = 'kpYaw = ' + str(params['kpYaw']) + '\n'
            lines = lines + [data]    
        elif line.startswith('kpPQR ='):
            temp = str(params['kpPQR'][0]) + ', ' + str(params['kpPQR'][1]) + ', ' + str(params['kpPQR'][2])
            data = 'kpPQR = ' + str(temp) + '\n'
            lines = lines + [data]
        else:
            lines = lines + [line]
    f.seek(0)
    f.writelines(lines)
    f.truncate()
```
